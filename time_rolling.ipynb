## Roll 1 minute to 5 minutes

### Prepare Data

```python
#Prepare 1 minutes timeseries data
#import yfinance as yf
import pandas as pd

symbols = ['1333.T', '1332.T', '1605.T']
#df = yf.download(symbols, group_by="ticker", interval="1m")
#df.iloc[:30].to_csv("./yfinance_1min_ex.csv")
```

```python
df_ = pd.read_csv("./yfinance_1min_ex.csv" ,header=[0,1], index_col=0, parse_dates=True)
```

```python
df_.iloc[:10]
```

### Roll data by pandas grouped by feature

Grouper of GroupedBy has freq option. It has many available frequencies for rolling time series data.

See details on https://pandas.pydata.org/docs/reference/api/pandas.Grouper.html

|Alias|Description|
|:--:|:--|
|B|business day frequency|
|C|custom business day frequency|
|D|calendar day frequency|
|W|weekly frequency|
|M|month end frequency|
|SM|semi-month end frequency (15th and end of month)|
|BM|business month end frequency|
|CBM|custom business month end frequency|
|MS|month start frequency|
|SMS|semi-month start frequency (1st and 15th)|
|BMS|business month start frequency|
|CBMS|custom business month start frequency|
|Q|quarter end frequency|
|BQ|business quarter end frequency|
|QS|quarter start frequency|
|BQS|business quarter start frequency|
|A, Y|year end frequency|
|BA, BY|business year end frequency|
|AS, YS|year start frequency|
|BAS, BYS|business year start frequency|
|BH|business hour frequency|
|H|hourly frequency|
|T, min|minutely frequency|
|S|secondly frequency|
|L, ms|milliseconds|
|U, us|microseconds|
|N|nanoseconds|


```python
freq = "5min"
```

```python
df = df_.copy()
df.columns = df.columns.swaplevel(0, 1)
df = df.iloc[3:].copy()
```

```python
rolled_opn = df["Open"].groupby(pd.Grouper(level=0, freq=freq)).first()
rolled_high = df["High"].groupby(pd.Grouper(level=0, freq=freq)).max()
rolled_low = df["Low"].groupby(pd.Grouper(level=0, freq=freq)).min()
rolled_cls = df["Close"].groupby(pd.Grouper(level=0, freq=freq)).last()
rolled_vlm = df["Volume"].groupby(pd.Grouper(level=0, freq=freq)).sum()
```

```python
rolled_df = pd.concat([rolled_opn, rolled_high, rolled_low, rolled_cls, rolled_vlm], axis=1, keys=["Open", "High","Low", "Close", "Volume"])
```

```python
rolled_df.columns = rolled_df.columns.swaplevel(0, 1)
rolled_df.sort_index(level=0, axis=1, inplace=True)
```

```python
rolled_df.iloc[:5]
```

### Check the Results

In case of minutes, data is rolled to freq - 1 from time of datetime.min % freq == 0

```python
df.columns = df.columns.swaplevel(0, 1)
df[symbols].iloc[:2]
```

```python
rolled_df.iloc[0:1]
```

#### data is rolled as expected

```python
expectations = {}
for symbol in symbols:
    expectation = {
        "Open": df[symbol]["Open"].iloc[0],
        "High":df[symbol][["Open", "High", "Low", "Close"]].iloc[:2].max().max(),
        "Low":df[symbol][["Open", "High", "Low", "Close"]].iloc[:2].min().min(),
        "Close":df[symbol]["Close"].iloc[1],
        "Volume":df[symbol]["Volume"].iloc[:2].sum()}
    expectations[symbol] = expectation
    print(f"{symbol}:{expectation}")
```

```python
for symbol in symbols:
    rolled_ohlc_df = rolled_df[symbol].iloc[0]
    expectation = expectations[symbol]
    for column in ["Open", "High", "Low", "Close", "Volume"]:
        if pd.notna(rolled_ohlc_df[column]):
            if expectation[column] != rolled_ohlc_df[column]:
                print(f"{rolled_ohlc_df[column]} != {expectation[column]} at {column} of {symbol}")
        else:
            if(pd.notna(expectation[column])):
                print(f"{rolled_ohlc_df[column]} != {expectation[column]} at {column} of {symbol}")
```

```python
#check next frame
frame_df = df[symbols].iloc[2:7]
frame_df
```

```python
rolled_df.iloc[1:2]
```

```python
expectations = {}

for symbol in symbols:
    expectation = {
        "Open": frame_df[symbol]["Open"].dropna().iloc[0],
        "High":frame_df[symbol][["Open", "High", "Low", "Close"]].max().max(),
        "Low":frame_df[symbol][["Open", "High", "Low", "Close"]].min().min(),
        "Close":frame_df[symbol]["Close"].dropna().iloc[-1],
        "Volume":frame_df[symbol]["Volume"].sum()}
    print(f"{symbol}: {expectation}")
```
